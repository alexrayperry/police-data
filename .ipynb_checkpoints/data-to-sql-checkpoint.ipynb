```python
import pandas as pd
import requests


```

```python
# Import & read CSV file

csv_data= 'ca_san_diego_2020_04-01.csv'


# Create dataframe based on CSV file

data = pd.read_csv(csv_data)

print(data)
```

```python
# Remove unwanted columns

df = pd.DataFrame(data)

df = df.drop(columns=['raw_row_number', 'type', 'reason_for_search', 'reason_for_stop', 'raw_action_taken', 'raw_subject_race_description'])

df.head(-20)


```

```python
values = {'date': "1/1/1999", 'time': "00:00:00", 'subject_age': 0, 'subject_race':"other", 'subject_sex': "unknown", 'arrest_made':"unknown",
          'citation_issued': "unknown", 'warning_issued': "unknown", 'outcome': "unknown", 'contraband_found': "unknown",
          'search_person':"unknown", 'search_vehicle': "unknown", 'search_basis': "other"}

df1 = df.fillna(value=values)

df1.head()

```

```python
# Change County to 100, Bulletin to 101, Unknown to 102 in service_area.


df2 = df1.replace({"County": "100", "Bulletin": "101", "Unknown": "102"})

df2.head()
```

```python
df2.astype({'date': 'datetime64', 'time': 'datetime64', 'service_area': 'int32', 'subject_race':'string', 'subject_sex':'string',
            'outcome': 'string'}).dtypes
```

```python
df3 = df2.sort_values(by=['date'])

df3.head()
```

```python

df3.to_csv('cleaned_version.csv',index=False)
```
